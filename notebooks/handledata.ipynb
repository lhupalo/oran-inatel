```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import *
import re
```

First, define your folder (in our case, the folder where the .txts form the simulations are stored).<br>

The cell below serves to load all .txt files into .pqt files, for each radius and sector.

```python
folder='./../results4x4/'
databases = []

# the args here are: (min radius, max radius, step=10). Adjust as your need according to the generated dataset
for raio in range(20, 30, 10):

    for sector in np.arange(1,2):    

        databases.append(pd.read_parquet(folder+'BC-random-positions-r=' + str(raio) + '-sector=' + str(sector) + '.pqt'))

    bigdata=pd.concat(databases)
    bigdata.reset_index(drop=True)
    bigdata.to_parquet(folder+'BC-random-positions-r=' + str(raio) + '.pqt')

    newdata=pd.read_parquet(folder+'BC-random-positions-r=' + str(raio) + '.pqt')
```

Now, the cell below serves to aggregate all .pqt files of each sector and radius on the final .pqt dataset used to train the ML models.

```python
databases = []

for raio in range(20, 30, 10):
    databases.append(pd.read_parquet(folder+'BC-random-positions-r=' + str(raio) + '.pqt'))

bigdata=pd.concat(databases)
bigdata.reset_index(drop=True)
bigdata.to_parquet(folder+'BC-random-positions_'+str(folder[12:15]) + '.pqt')

newdata=pd.read_parquet(folder+'BC-random-positions_'+str(folder[12:15]) + '.pqt')
print(newdata)
```
